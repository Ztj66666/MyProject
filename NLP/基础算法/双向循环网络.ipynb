```python
import torch
from torch import nn
import dltools
```

```python
batch_size,num_steps = 32,25
train_iter,vocab = dltools.load_data_time_machine(batch_size,num_steps)
```

```python
vocab_size,num_hiddens , num_layers = len(vocab),256,2
device = dltools.try_gpu()
num_inputs = vocab_size
rnn_layer = nn.RNN(num_inputs,num_hiddens,num_layers,bidirectional=True,device=device)

model = dltools.RNNModel(rnn_layer,vocab_size)
model = model.to(device)
```

```python
num_epoches ,lr=500,1
dltools.train_ch8(model,train_iter,vocab,lr,num_epoches,device)
```

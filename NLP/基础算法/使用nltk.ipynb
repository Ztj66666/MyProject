```python
[word for word in tokenized_word if word not in punctuation]
```

```python
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
```

```python
text = "'Is not that rather a large thing to expect us to begin upon?'said Filby, an argumentative person with red hair."
tokenized_text = word_tokenize(text)
tokenized_text = [word for word in tokenized_text if word not in punctuation and word not in stop_words]
tokenized_text
```

```python
import nltk
```

```python
tokenized_text = nltk.word_tokenize(text)
tokenized_text = [word for word in tokenized_text if word not in punctuation and word not in stop_words]
word_frequency = nltk.FreqDist(w.lower() for w in tokenized_text)
word_frequency
```

```python
word_frequency.plot()
```

```python
word_frequency.plot(2,cumulative=True)
```
